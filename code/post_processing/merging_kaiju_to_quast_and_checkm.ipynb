```python
import pandas as pd
```

```python
quast_df = pd.read_csv('quast_mapping_file.tsv',sep="\t",index_col=0)
quast_df.head(1)
```

```python
checkm_df = pd.read_csv('checkm_mapping_file.tsv',sep="\t",index_col = 0)
checkm_df.head(1)
```

```python
df = checkm_df.join(quast_df)
df.head(1)
```

```python
df.columns
```

```python
df.drop(columns=['N50 (contigs)','N50 (scaffolds)','Total length'],inplace=True)
df.head(1)
```

```python
kaiju = pd.read_csv('kaiju_mapping_file.tsv',sep="\t",index_col=0)
kaiju.head(1)
```

```python
df = df.join(kaiju)
df.head(1)
```

```python
df.columns
```

```python
new_order = ['top_taxon', 'percent', 'reads', 'Completeness', 'Contamination', 
             'Genome size', '# contigs', 'GC (%)', 'N50', 'N90', 'auN', 'L50', 'L90', 'remaining']
```

```python
df = df[new_order]
df.head(1)
```

```python
df.rename(columns={'percent': 'kaiju percent', 'reads': 'kaiju reads'}, inplace=True)
df.head(1)
```
